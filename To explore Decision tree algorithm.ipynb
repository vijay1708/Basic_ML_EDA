```python
#importing all libraries required for this project
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
iris.data
```

```python
#feature names present in the dataset
iris.feature_names
```

```python
iris.target
```

```python
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print("Data imported successfully")
df
```

```python
df['target']=iris.target
df
```

```python
sns.scatterplot(x= 'petal width (cm)', y= 'petal length (cm)', data=df)
```

```python
df.apply(lambda x:sum(x.isnull()))  #checking for null values
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('target',axis=1)
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) #splitting data into training and test sets
```

```python
#Training our algorithm
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn import tree
feat_names=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
classes=['setosa', 'versicolor', 'virginica']

fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(dtree,
               feature_names = feat_names, 
               class_names=classes,
               filled = True);
```
